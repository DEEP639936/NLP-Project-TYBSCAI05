```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import re
import string 
import nltk
```

```python
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
text = """
Machine Learning is one of the most important fields of Artificial Intelligence.
It enables computers to learn from data without being explicitly programmed.
Today, Machine Learning is used in healthcare, banking, education, e-commerce,
cybersecurity, self-driving cars, and recommendation systems.

Students study Machine Learning because it helps solve real-world problems.
However, before applying any machine learning algorithm,
the textual data must be cleaned and preprocessed.
Text preprocessing improves the quality of data by removing unnecessary
symbols, stopwords, punctuation, numbers, and extra spaces.
"""
```

```python
print("Original Corpus")
print(text)
```

```python
sentences = sent_tokenize(text)

print("Sentence Tokenization")
print("-"*50)

for i, sentence in enumerate(sentences, start=1):
    print(f"Sentence {i}: {sentence}")
```

```python
words = word_tokenize(text)

print("Word Tokenization")
print("-"*50)

print(words)
```

```python
lower_words = [word.lower() for word in words]

print("Lowercase Words")
print("-"*50)

print(lower_words)
```

```python
words_no_punctuation = [
    word for word in lower_words
    if word not in string.punctuation
]

print("After Removing Punctuation")
print("-"*50)

print(words_no_punctuation)
```

```python
words_no_numbers = [
    word for word in words_no_punctuation
    if word.isalpha()
]

print("After Removing Numbers")
print("-"*50)

print(words_no_numbers)
```

```python
stop_words = set(stopwords.words("english"))

filtered_words = [
    word for word in words_no_numbers
    if word not in stop_words
]

print("After Stopword Removal")
print("-"*50)

print(filtered_words)
```

```python
stemmer = PorterStemmer()

stemmed_words = [
    stemmer.stem(word)
    for word in filtered_words
]

print("After Stemming")
print("-"*50)

print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [
    lemmatizer.lemmatize(word)
    for word in filtered_words
]

print("After Lemmatization")
print("-"*50)

print(lemmatized_words)
```

```python
final_corpus = " ".join(lemmatized_words)

print("Final Clean Corpus")
print("-"*50)

print(final_corpus)
```

```python
with open("final_corpus.txt", "w", encoding="utf-8") as file:
    file.write(final_corpus)

print("Corpus Saved Successfully!")
```
