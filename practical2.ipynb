```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

import nltk
import string
import re

nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
text = """
Natural Language Processing is an exciting field of Artificial Intelligence.
It allows computers to understand, process, and generate human language.
"""
```

```python
sentences = sent_tokenize(text)

print("Sentence Tokenization")
print("-"*50)

for sentence in sentences:
    print(sentence)
```

```python
words = word_tokenize(text)

print("Word Tokenization")
print("-"*50)

print(words)
```

```python
characters = list(text)

print("Character Tokenization")
print("-"*50)

print(characters)
```

```python
lower_text = text.lower()

print("Lowercase")
print("-"*50)

print(lower_text)
```

```python
text_without_punctuation = lower_text.translate(
    str.maketrans("", "", string.punctuation)
)

print("Without Punctuation")
print("-"*50)

print(text_without_punctuation)
```

```python
text_without_numbers = re.sub(r"\d+", "", text_without_punctuation)

print("Without Numbers")
print("-"*50)

print(text_without_numbers)
```

```python
normalized_text = re.sub(r"\s+", " ", text_without_numbers).strip()

print("Normalized Text")
print("-"*50)

print(normalized_text)
```

```python
final_tokens = word_tokenize(normalized_text)

print("Final Tokens")
print("-"*50)

print(final_tokens)
```
